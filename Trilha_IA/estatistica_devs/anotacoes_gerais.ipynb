## Medidas de posição - Média,Mediana e Moda    (mean)  (median)    (mode)
## Medidas de dispersão Variância, Desvio Padrão e Coeficiente de Variação:
#### variância : média dos quadrados das diferenças de cada valor (var())
#### Desvio Padrão : raiz quadrada da variância (std())
#### Coeficiente de Variação : Desvio padrão dividido pela média (%) (std()/mean()*100)
## Medidas de forma - Assimetria e Curtose
#### Assimetria : Grau e direção da distorção da distribuição em relação a média 
#### Curtose :  Mede o pico (pontuação) da distribuição

## Correlação
#### mede a relação entre as duas variaveis. Se tem uma associação positiva ou negativa (as duas aumentam juntas, ou uma aumenta e a outra diminui) -> prêve com maior acurácia a relação entre ela. Pode ser também sem associação.
### coeficiente de Correlação
#### Coeficiente de PEARSON - Mede a correlação linear entre duas variaveis -> -1 (negativa perfeita) +1 (positiva perfeita) 0 (nenhuma correlação)
#### Coeficiente de SPEARMAN - relação monotônica - não necessáriamente linear clara ou quando não são numericamente escalonáveis, não númericas (-1,0,+1)
