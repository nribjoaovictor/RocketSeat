```python
pd.__version__
```

```python
dict_medidas = {'idade': [15,18,25,18,55,58,60,80], 'altura':[160,162,165,172,174,174,176,180]}
```

```python
df_medidas = pd.DataFrame.from_dict(dict_medidas)
```

```python
df_medidas
```

## MEDIDAS RESUMO


### Medidas de Posição

```python
# média
df_medidas["idade"].mean()
```

```python
df_medidas["idade"].median()
```

```python
df_medidas["idade"].mode()
```

### Medidas de Dispersão

```python
# Variância
df_medidas.idade.var()
```

```python
# Desvio padrão - standard Variation
df_medidas.idade.std()
```

```python
# Coeficiente de variação - idade (%)
df_medidas.idade.std() / df_medidas.idade.mean() *100
```

```python
# Coeficiente de variação - altura (%)
df_medidas.altura.std() / df_medidas.altura.mean() *100
```

### Medidas de Forma

```python
# Assimetria - idade
df_medidas.idade.skew()
```

```python
# Assimetria - altura
df_medidas.altura.skew()
```

```python
# curtose - idade
df_medidas.idade.kurtosis()
```

```python
# curtoses - altura
df_medidas.altura.kurtosis()
```

```python
# obter medidas estatísticas 
df_medidas.idade.describe()
```

### coeficiente de Correlação - correlaçao de variáveis

```python
df_medidas.corr(method='pearson') 
```

```python
# correlação entre duas variaveis
df_medidas.idade.corr(df_medidas.altura)
```

```python
# correlação de Spearman
df_medidas.corr(method='spearman')
```

## Criar um dataframe de vendas


```python
dict_vendas = {'categorias':['masculino','feminino','infantil','casa'],'valor':[400000,700000,250000,580000],'quantidade':[3000,5000,1500,2500]}
```

```python
df_vendas = pd.DataFrame.from_dict(dict_vendas)
df_vendas
```

```python
# Gráfico de barras verticais
df_vendas.plot.bar(x='categorias',y='valor')
```

```python
#gráfico de barra vertical valor ordenado
df_vendas.sort_values('valor', ascending=False).plot.bar(x='categorias',y='valor')
```

```python
# Gráfico de barra horizontal - quantidade
df_vendas.plot.barh(x='categorias',y='quantidade')
```

```python
# gráfico de barras horizontal quantidade ordenado - ranking
df_vendas.sort_values('quantidade').plot.barh(x='categorias',y='quantidade')
```

```python
# Gráfico de dispersão
df_medidas.plot.scatter(x='idade',y='altura')
```

```python
# gráfico de boxPlot ou diagrama de caixa - idade
df_medidas.idade.plot.box()
```

nesse caso, não se tem outliers

```python
df_medidas.altura.plot.box()
```

```python
# gerar dataFrame com dados temporais
dict_faturamento = {'data_ref':['2019-01-01','2020-01-01','2021-01-01','2022-01-01','2023-01-01'],'valor':[400000,800000,500000,800000,900000]}
dict_faturamento
```

```python
dt_faturamento.info()
```

```python
dt_faturamento['data_ref'] = pd.to_datetime(dt_faturamento['data_ref'])
```

```python
#Gráfico de linhas
dt_faturamento.plot.line(x='data_ref',y='valor')
```
